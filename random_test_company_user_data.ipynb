```python

import urllib.request
import bs4

url = "file:///C:/Users/byeonghooh/Documents/ETRI/1.%20%EB%8D%B0%EC%9D%B4%ED%84%B0%20%EB%B6%84%EC%84%9D%20%EB%B0%8F%20%ED%86%B5%EA%B3%84%20DB%20%EA%B5%AC%EC%B6%95/html_path/imhyeongyeong.html"
html = urllib.request.urlopen(url)

bsObj = bs4.BeautifulSoup(html, "html.parser", from_encoding = 'utf-8')

#print(html.read())
print(bsObj)
```
